```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import albumentations as a
```

```python
img_path = glob('../data/selected_data/all/*png')
```

```python
img_path = sorted(img_path)
```

```python
def read_img(path):
    img = cv2.imread(path)
    img = img[..., ::-1]
    return img
```

```python
def image_cnt_crop(img, H=1900, W=1900):
    crop_img = a.center_crop(img, H, W)
    return crop_img
```

```python
for path in img_path:
    img = read_img(path)
    crop_img = image_cnt_crop(img)
    cv2.imwrite(path, crop_img)
```

```python
img_list = list()

for path in img_path:
    img = read_img(path)
    img_list.append(img)
```

```python
np.array(img_list).shape
```

```python
idx = 1
H, W = (1800, 1800)
fig = plt.figure(figsize=(12, 12))
plt.title(img_path[idx])
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.imshow(img_list[idx])
ax2.imshow(a.center_crop(img_list[idx], H, W))
plt.show()
```
